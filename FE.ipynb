```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
pd.pandas.set_option("display.max_columns",None)
data = pd.read_csv("train.csv")
data.head()
```

## categorical Feature

```python
categorical_fea = [feature for feature in data.columns if data[feature].dtype == 'O']

#print(len(Categorical_fea), len(Numerical_fea))
```

```python
categorical_fea_nan = [feature for feature in categorical_fea if data[feature].isnull().sum() > 0]
for feature in categorical_fea_nan:
    print(feature, f"have {round(np.mean(data[feature].isnull())*100,4)} % missing value")
```

```python
df = data.copy()
for feature in categorical_fea_nan:
    df[feature] = np.where(df[feature].isnull(), 'Missing',df[feature])
    print(feature, f"have {round(np.mean(df[feature].isnull())*100,4)} % missing value")
df
```

## Numerical 

```python
numerical_fea = [feature for feature in data.columns if data[feature].dtype != 'O' and feature not in ['Id', 'SalePrice']]
numerical_fea_nan = [feature for feature in numerical_fea if df[feature].isnull().sum() > 0]
for feature in numerical_fea_nan:
    print(feature, f"have {round(np.mean(df[feature].isnull())*100,2)} % missing value")
```

```python
for feature in numerical_fea_nan:
    df[feature+"_nan"] = np.where(df[feature].isnull(), 1, 0)
    df[feature] = np.where(df[feature].isnull(),df[feature].median(),df[feature])
    print(feature, f"have {round(np.mean(df[feature].isnull())*100,2)} % missing value")
```

## Year Balancing

```python
year_fea = [feature for feature in data.columns if "Year" in feature or 'Yr' in feature]
for feature in year_fea[:-1]:
    df[feature] = df['YrSold'] - df[feature]
df[year_fea[:-1]].head()
```

## Skewed Numerical

```python
skewed_num = ["LotFrontage", '1stFlrSF', 'GrLivArea', 'SalePrice']
for feature in skewed_num:
    df[feature] = np.log(df[feature])
df.head()
```

## Rare Category

```python
for feature in categorical_fea:
    temp = df[feature].value_counts()/len(data)
    temp_df = temp[temp>0.01].index
    df[feature] = np.where(df[feature].isin(temp_df), df[feature], "Rare_val")
df.head(50)
```

```python
for feature in categorical_fea:
    label_value = df.groupby(feature)['SalePrice'].mean().sort_values().index
    print(label_value)
    labeled_value = {i:k for k, i in enumerate(label_value)}
    df[feature] = df[feature].map(labeled_value)
df.head(50)
```

```python
from sklearn.preprocessing import MinMaxScaler
feature_scale = [feature for feature in df.columns if feature not in ['Id','SalePrice']] 
scaler = MinMaxScaler()
scaler.fit(df[feature_scale])
scaled_val = scaler.transform(df[feature_scale])
scaled_val
```

```python
data1 = pd.concat([data[['Id','SalePrice']], pd.DataFrame(scaled_val,columns = feature_scale)])
data1.head()
```
